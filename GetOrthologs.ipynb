```python
import pandas as pd
```

```python
df_AE = pd.read_table('OutputFiles/SQL/AvE_trans.out', names=['qseqid', 'sseqid', 'qlen', 'slen', 'bitscore', 'evalue', 'pident', 'nident', 'length', 'qcovs', 'qstat', 'qend', 'sstart', 'send', 'scov'])
df_EA = pd.read_table('OutputFiles/SQL/EvA_trans.out', names=['qseqid', 'sseqid', 'qlen', 'slen', 'bitscore', 'evalue', 'pident', 'nident', 'length', 'qcovs', 'qstat', 'qend', 'sstart', 'send', 'scov'])
```

```python
idx = df_AE.groupby(['qseqid'])['bitscore'].transform(max) == df_AE['bitscore']
max_df_AE = df_AE[['qseqid', 'sseqid', 'bitscore']].loc[idx]

idx = df_EA.groupby(['qseqid'])['bitscore'].transform(max) == df_EA['bitscore']
max_df_EA = df_EA[['qseqid', 'sseqid', 'bitscore']].loc[idx]
```

```python
df_homologs = pd.DataFrame(columns=['g1', 'g2', 'relationship', 'method'])
```

```python
for i, row in max_df_AE.iterrows():
    if row.qseqid in max_df_EA.loc[max_df_EA['qseqid'] == row.sseqid]['sseqid'].values:
        df_homologs.loc[len(df_homologs)] = [row.qseqid, row.sseqid, 'Orthologs', 'BDBH']
    
```
